```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
file_path = "listings.csv"   
df = pd.read_csv(file_path)
df.head()
```

```python
missing = df.isnull().sum() / len(df) * 100
missing.sort_values(ascending=False)
```

```python
columns_to_drop = [
    'listing_url', 'scrape_id', 'last_scraped', 
    'picture_url', 'host_url', 'host_thumbnail_url',
    'host_picture_url'
]

df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True)
df['price'] = df['price'].astype(float)
```

```python
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
df['days_since_last_review'] = (pd.Timestamp.today() - df['last_review']).dt.days
```

```python
df['price_category'] = pd.cut(
    df['price'],
    bins=[0, 100, 250, df['price'].max()],
    labels=['Low', 'Medium', 'High']
)
```

```python
upper_limit = df['price'].quantile(0.99)
df = df[df['price'] <= upper_limit]
```

```python
df = pd.get_dummies(
    df,
    columns=['room_type', 'neighbourhood_group'],
    drop_first=True
)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv("cleaned_airbnb_listings.csv", index=False)
```

```python
# Analysis and Visualisation
```

```python
# Insight 1: Price Distribution is Right-Skewed (Luxury Listings Drive High Prices)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.histplot(df['price'], bins=50)
plt.title("Distribution of Listing Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Insight 2: Availability Strongly Influences Number of Reviews
```

```python
plt.figure()
sns.scatterplot(x='availability_365', y='number_of_reviews', data=df)
plt.title("Availability vs Number of Reviews")
plt.xlabel("Availability (days)")
plt.ylabel("Number of Reviews")
plt.show()

df[['availability_365','number_of_reviews']].corr()
```

```python
# Insight 3: Private Rooms Are Significantly Cheaper Than Entire Properties
```

```python
df['room_type'] = 'Entire home'

df.loc[df['room_type_Private room'] == 1, 'room_type'] = 'Private room'
df.loc[df['room_type_Shared room'] == 1, 'room_type'] = 'Shared room'
df.loc[df['room_type_Hotel room'] == 1, 'room_type'] = 'Hotel room'
```

```python
plt.figure()
sns.boxplot(x='room_type', y='price', data=df)
plt.title("Price Distribution by Room Type")
plt.xticks(rotation=45)
plt.show()
```

```python
# Insight 4: High-Review Listings Tend to Have Higher Prices (Reputation Effect)
```

```python
plt.figure()
sns.regplot(x='number_of_reviews', y='price', data=df, scatter_kws={'alpha':0.3})
plt.title("Reviews vs Price (Reputation Effect)")
plt.xlabel("Number of Reviews")
plt.ylabel("Price")
plt.show()

df[['number_of_reviews','price']].corr()
```
